```python
class Cafe :
    
    def __init__(self) :
        self.table_List = list()
        self.menu_List = list()
        self.dagi_List = list()
        self.name = ""
    
    def get_Name(self) :
        return self.name
    
    def set_Name(self) :
        name = input("가게 이름 : ")
        self.name = name
        
    def get_Table(self) :
        return self.table_List
    
    def set_Table(self) :
        n = int(input("테이블 개수 = "))
        for i in range(1, n+1) :
            self.get_Table().append([i, 0, []])
    
    def init(self) :
        self.table_List = []
        self.menu_List = []
        self.dagi_List = []
        self.name = ""
        
        
# 매뉴 ------------------------------------------------------------------

    def get_Menu(self) :
        return self.menu_List
    
    def selet_Menu(self) :
        print("-"*20)
        print("번호    이름     가격")
        for menu in self.get_Menu() :
            print("%s %7s %10s원" %(menu[0],menu[1],menu[2]))
        print("-"*20)
        
    def set_Menu(self) :
        print("-- < %s > --" %self.get_Name())
        print("-"*20)
        print("매뉴 추가 : 0 = 종료")
        print("-"*20)
        
        index = 0
        while True :
            menu = input("매뉴 이름 : ")
            if menu == "0" :
                self.selet_Menu()
                break
            
            else :
                price = int(input("가격 설정 : "))
                index += 1
                self.get_Menu().append([index, menu, price])
                print("= 매뉴추가 되었습니다 =\n")
                
# 주문 --------------------------------------------------------------

    def table_Check(self) :
        for table in self.get_Table() :
            if table[1] == 0 :
                table[1] = 1
                
                domain = len(self.get_Menu())
                self.selet_Menu()
                
                while True :
                    num = int(input("매뉴 선택하세요 (종료하려면 0입력)"))
                    if str(num) == "0" :
                        break
                        
                    elif num >= 1 and num <= domain :
                        table[2].append(num)
                break
                       
                
    def check_Out(self) :
        tableNumber = int(input("결제 - 테이블 번호 = "))
        for table in self.get_Table() :
            if table[0] == tableNumber :
                menu_Dmain = len(table[2])
                self.select_Receipt(table[2], menu_Dmain)
                table[1] = 0
                table[2] = []
                
    
    def select_Receipt(self, table, menu_Dmain) : # 영수증 출력
        print("====== 영수증 ======")
        print("---" + self.get_Name() + "---")
        print("요리명      금액")
        num = 0
        tot = 0
        for i in table :
            if i == self.get_Menu()[num][0] :
                tot += self.get_Menu()[num][2]
                print("%s %d" %(self.get_Menu()[num][1],self.get_Menu()[num][2]))
                num += 1
        print("-------------------")
        print("총 결제금액 = %d\n" % tot)
                
            
    
    
    def add_cust(self) :
        if self.full_Check() == True :
            
            dagi_number = len(self.get_dagi())
            print("자리가 없습니다. 앞에 대기중인 손님 %d명 "%dagi_number)
            self.get_dagi().append("cust")
            
        else :
            if self.get_LenDagi() != 0 :
                del self.get_dagi()[0]
                self.table_Check()
            else :
                self.table_Check()
        
    def full_Check(self) :
        count = 0
        for i in self.get_Table() :
            if i[1] == 1 :
                count += 1
        
        if count == len(self.get_Table()) :
            return True

            
    
# 대기 --------------------------------------------------------------

    def get_dagi(self) :
        return self.dagi_List  
    
    def get_LenDagi(self) :
        return len(self.get_dagi())

    
#--------------------------------------------------------------------

    def seting(self) :

        while True :
            print()
            print("="*20)
            print("1. 가게 Open\n2. 테이블 조회\n3. 가게 변경\n4. 고객 추가\n5. 고객 결제\n0. 작업 종료")
            print("="*20)
            num = int(input("선택 = "))
            if num == 1 :
                self.set_Name()
                self.set_Menu()
                self.set_Table()

            elif num == 2 :
                print("\n>-------- 테이블 현황 --------<\n")
                for table in self.get_Table() :
                    if table[1] == 1 :
                        print("   - 테이블[%d] (%s)" %(table[0],"사용중..."))
                    else :
                        print("   - 테이블[%d] (%s)" %(table[0],"비어 있음..."))
                
                print("\n>-------- 개기줄 현황 --------<\n")
                
                if len(self.get_dagi()) == 0 :
                    print("   - 대기 없음 ")
                    
                else :
                    print("   - ",self.get_dagi())

            elif num == 3 :
                self.init()
                self.set_Name()
                self.set_Menu()
                self.set_Table()
                
            elif num == 4 :
                self.add_cust()
                
            elif num == 5 :
                self.check_Out()
                
            elif num == 0 :
                break
```

```python
a = Cafe()
a.seting()
```
