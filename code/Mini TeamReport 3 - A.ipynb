```python
class Cafe :
    
    def __init__(self) : #클래스 호출할 때마다 호출되는 거'__init__'
        self.table_List = list() #list함수를 table_List로 변수 설정 값이 담겨있지 않다.(빈 리스트)
        self.menu_List = list()
        self.stay_List = list()
        self.name = ""
    
    def get_Name(self) :
        return self.name
    
    def set_Name(self) :
        name = input("가게 이름 : ")
        self.name = name
        
    def get_Table(self) :
        return self.table_List
    
    def set_Table(self) :
        n = int(input("테이블 개수 = "))
        for i in range(1, n+1) :
            self.get_Table().append([i, 0, []])
    
    def init(self) :  # clear
        self.table_List = []
        self.menu_List = []
        self.stay_List = []
        self.name = ""
        
        
# 메뉴 ------------------------------------------------------------------

    def get_Menu(self) :
        return self.menu_List
    
    def select_Menu(self) :
        print("-"*30)
        print("번호    이름     가격") # \t
        for menu in self.get_Menu() :
            print("%s %7s %10s원" %(menu[0],menu[1],menu[2]))
        print("-"*30)
        
    def set_Menu(self) :
        print("-- < %s > --" %self.get_Name())
        print("-"*30)
        print("메뉴 추가 : 0 = 종료")
        print("-"*30)
        
        index = 0
        while True :
            menu = input("메뉴 이름 : ")
            if menu == "0" :
                self.select_Menu()
                break
            
            else :
                price = int(input("가격 설정 : "))
                index += 1
                self.get_Menu().append([index, menu, price])
                print("= 메뉴 추가 되었습니다 =\n")
                
# 주문 --------------------------------------------------------------

    def table_Check(self) :
        for table in self.get_Table() :
            if table[1] == 0 :
                table[1] = 1
                
                domain = len(self.get_Menu())
                self.select_Menu()
                
                while True :
                    num = int(input("메뉴 선택하세요 (종료하려면 0입력)"))
                    if str(num) == "0" :
                        break
                        
                    elif num >= 1 and num <= domain :
                        table[2].append(num)
                break
                       
                
    def check_Out(self) :
        tableNumber = int(input("결제 - 테이블 번호 = "))
        for table in self.get_Table() :
            if table[0] == tableNumber :
                menu_Dmain = len(table[2])
                self.select_Receipt(table[2], menu_Dmain)
                table[1] = 0
                table[2] = []
                
    
    def select_Receipt(self, table, menu_Dmain) : # 영수증 출력
        print("====== 영수증 ======")
        print("---" + self.get_Name() + "---")
        print("메뉴      금액")
        num = 0
        tot = 0
        for i in table :
            if i == self.get_Menu()[num][0] :
                tot += self.get_Menu()[num][2]
                print("%s %d" %(self.get_Menu()[num][1],self.get_Menu()[num][2]))
                num += 1
        print("-------------------")
        print("총 결제금액 = %d\n" % tot)
                
            
    
    
    def add_cust(self) :
        if self.full_Check() == True : #비워져있는 테이블이 있냐 없냐
            
            stay_number = len(self.get_stay())
            print("자리가 없습니다. 앞에 대기중인 손님 %d명 "%stay_number)
            cust = input("대기명을 입력하세요 ex)한소희, 꼬부기, 박보검...>>>")
            self.get_stay().append(cust)
            
            if stay_number > 8:
                print("죄송합니다. 해당 대기인원을 받을 수 없습니다. 다음에 저희 매장을 찾아주시길 바랍니다.")
            
        else : #빈테이블이 있다
            if self.get_LenStay() != 0 :
                del self.get_stay()[0]
#                 print("다음 손님은", self.get_stay(), "입니다. 잠시만 기다려주세요.")
                self.table_Check() #손님을 앉히는 과정
            else :
                self.table_Check()
        
    def full_Check(self) :
        count = 0
        for i in self.get_Table() :
            if i[1] == 1 :
                count += 1
        
        if count == len(self.get_Table()) : #테이블 수가2,
            return True




            
    
# 대기 --------------------------------------------------------------


    def get_stay(self) :
        print("다음 손님은", self.get_stay(), "입니다. 잠시만 기다려주세요.")
        return self.stay_List  
    
    def get_LenStay(self) :
        return len(self.get_stay())

    
#--------------------------------------------------------------------

    def seting(self) :

        while True :
            print()
            print("="*30)
            print("1. 가게 Open\n2. 테이블 조회\n3. 가게 변경\n4. 고객 추가\n5. 고객 결제\n0. 작업 종료")
            print("="*30)
            num = int(input("선택 = "))
            if num == 1 :
                self.set_Name()
                self.set_Menu()
                self.set_Table()

            elif num == 2 :
                print("\n>-------- 테이블 현황 --------<\n")
                for table in self.get_Table() :
                    if table[1] == 1 :
                        print("   - 테이블[%d] (%s)" %(table[0],"사용중..."))
                    else :
                        print("   - 테이블[%d] (%s)" %(table[0],"비어 있음..."))
                
                print("\n>-------- 대기줄 현황 --------<\n")
                
                if len(self.get_stay()) == 0 :
                    print("   - 대기 없음 ")
                    
                else :
                    print("   - ",self.get_stay())

            elif num == 3 :
                self.init()
                self.set_Name()
                self.set_Menu()
                self.set_Table()
                
            elif num == 4 :
                self.add_cust()
                
            elif num == 5 :
                self.check_Out()
                
            elif num == 0 :
                break
```
