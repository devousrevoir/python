```python
class Cafe :
    
    def __init__(self) :
        self.table_List = list()
        self.menu_List = list()
        self.stay_List = list()
        self.name = ""
    
    def get_Name(self) :
        return self.name
    
    def set_Name(self) : 
        name = input("상호명 : ")
        self.name = name
        
    def get_Table(self) :
        return self.table_List
    
    def set_Table(self) :
        n = int(input("테이블 수 = "))
        for i in range(1, n+1) :
            self.get_Table().append([i, 0, []])
    
    def init(self) :
        self.table_List = []
        self.menu_List = []
        self.stay_List = []
        self.name = ""


    def get_Menu(self) :
        return self.menu_List
    
    def select_Menu(self) :
        print("-"*40)
        print("번호     이름     가격")
        for menu in self.get_Menu() :
            print("%s %7s %10s원" %(menu[0],menu[1],menu[2]))
        print("-"*40)
        
    def set_Menu(self) :
        print("-- < %s > --" %self.get_Name())
        print("-"*40)
        print("메뉴 추가 : 0 = 종료")
        print("-"*40)
        
        index = 0
        while True :
            menu = input("메뉴 : ")
            if menu == "0" :
                self.select_Menu()
                break
            
            else :
                price = int(input("금액 설정 : "))
                index += 1
                self.get_Menu().append([index, menu, price])
                print("= 메뉴 추가 되었습니다 =\n")


    def table_Check(self) :
        for table in self.get_Table() :
            if table[1] == 0 :
                table[1] = 1
                
                domain = len(self.get_Menu())
                self.select_Menu()
                
                while True :
                    num = int(input("메뉴 선택하세요 (종료하려면 0입력)"))
                    if str(num) == "0" :
                        break
                        
                    elif num >= 1 and num <= domain :
                        table[2].append(num)
                break
                       
                
    def check_Out(self) :
        tableNumber = int(input("결제 - 테이블 번호 = "))
        for table in self.get_Table() :
            if table[0] == tableNumber :
                menu_Dmain = len(table[2])
                self.select_Receipt(table[2], menu_Dmain)
                table[1] = 0
                table[2] = []
                
    
    def select_Receipt(self, table, menu_Dmain) : 
        print("======= 영수증 =======")
        print("---" + self.get_Name() + "---")
        print("메뉴              금액")
        num = 0
        tot = 0
        []
        for i in table :
            if i == self.get_Menu()[num][0] :
                tot += self.get_Menu()[num][2]
                print("%s %d" %(self.get_Menu()[num][1],self.get_Menu()[num][2]))
                num += 1
        print("------------------------")
        print("총 결제금액 = %d\n" % tot)
        print("이용해주셔서 감사합니다.") # 앞의 문장을 사용하고 싶으면 \n을 사용해서 아예 총 결제금액 옆에 붙여야 하는지?
                
            
    
    
    def add_cust(self) :
        if self.full_Check() == True :
            
            stay_number = len(self.get_stay())
            print("자리가 없습니다. 앞에 대기중인 손님 %d명 "%stay_number)
            self.get_stay().append("cust")
            
        else :
            if self.get_LenStay() != 0 :
                del self.get_stay()[0]
                self.table_Check()
            else :
                self.table_Check()
        
    def full_Check(self) :
        count = 0
        for i in self.get_Table() :
            if i[1] == 1 :
                count += 1
        
        if count == len(self.get_Table()) :
            return True


    def get_stay(self) :
        return self.stay_List  
    
    def get_LenStay(self) :
        return len(self.get_stay())


    def seting(self) :

        while True : # 모르고 숫자 6을 두 번 입력했는데 생성 가능함. 아예 존재하지 않는 번호입니다. 문구 사용 가능한지 ?? 
            print()
            print("="*30)
            print("1. 가게 Open\n2. 테이블 조회\n3. 고객 추가\n4. 고객 결제\n5. 가게 변경\n6. 작업 종료")
            print("="*30)
            num = int(input("선택 = "))
            if num == 1 :
                self.set_Name()
                self.set_Menu()
                self.set_Table()

            elif num == 2 :
                print("\n>-------- 테이블 현황 --------<\n")
                for table in self.get_Table() :
                    if table[1] == 1 :
                        print("   - 테이블[%d] (%s)" %(table[0],"사용중..."))
                    else :
                        print("   - 테이블[%d] (%s)" %(table[0],"비어있음..."))
                
                print("\n>-------- 대기줄 현황 --------<\n")
                
                if len(self.get_stay()) == 0 :
                    print("   - 대기 없음 ")
                    
                else :
                    print("   - ",self.get_stay())

            elif num == 3 :
                self.init()
                self.set_Name()
                self.set_Menu()
                self.set_Table()
                
            elif num == 4 :
                
                self.add_cust()
                
            elif num == 5 :
                self.check_Out()
                
            elif num == 6 :
                break
```
